```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
museum_data = pd.read_csv("/content/museum-visitors.csv")
museum_data.head()
```

```python
#  Line Chart
plt.figure(figsize=(12,6))
plt.title("Monthly Visitors to Avila Adobe")
sns.lineplot(data=museum_data['Avila Adobe'])
plt.xlabel("Date")
```

```python
#  Bar Chart
plt.figure(figsize=(8, 6))
sns.barplot(x=museum_data['Avila Adobe'], y=museum_data.index)
plt.xlabel("")
plt.title("Monthly Visitors to Avila Adobe")
```

```python
#  Scatter Plot
sns.scatterplot(x=museum_data['Month'], y=museum_data['Avila Adobe'])
```

```python
#  Pie Chart
plt.figure(figsize =(10, 7))
plt.pie(museum_data['Avila Adobe'])
```

```python
#  Histogram
sns.histplot(museum_data['Avila Adobe'])
```

```python
#  Boxplot
plt.boxplot(museum_data['Avila Adobe'])
```

```python
# KDE plot 
sns.kdeplot(data=museum_data['Avila Adobe'], shade=True)
```

```python
# KDE2 plot 

sns.kdeplot(data=museum_data, x='Avila Adobe')

plt.title("Distribution of Avila Adobe")
```
