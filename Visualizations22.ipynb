```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
maths_data = pd.read_csv("/content/Maths.csv")
maths_data.head()
```

```python
#  Line Chart
plt.figure(figsize=(20,10))
plt.title("Absences")
sns.lineplot(data=maths_data['absences'])
plt.xlabel("absences")
```

```python
#  Bar Chart
plt.figure(figsize=(20,10))
sns.barplot(x=maths_data['Mjob'], y=maths_data.index)
plt.xlabel("Mathores Job")
plt.title("maths_data")
```

```python
#  Scatter Plot
plt.figure(figsize=(20,10))
plt.title("Absences")
plt.xlabel("Age")
sns.scatterplot(x=maths_data['age'], y=maths_data['school'])
```

```python
#  Pie Chart
plt.figure(figsize=(20,10))
maths_data['sex'].value_counts().head(10).plot.pie()
plt.gca().set_aspect('equal')
plt.title("Sex of student's school")
```

```python
#  Histogram
plt.figure(figsize=(20,10))
plt.title("health students")
sns.histplot(maths_data['health'])
plt.xlabel("number")

```

```python
#  Boxplot
plt.figure(figsize=(20,10))
plt.title("Freetime of students")
plt.boxplot(maths_data['freetime'])
plt.xlabel("hours")
```

```python
# subplot
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
sns.barplot(x=maths_data['Mjob'], y=maths_data.index)
plt.xlabel("Mathores Job")

plt.subplot(2,2,2)
sns.histplot(maths_data['health'])

plt.subplot(2,2,3)
maths_data['sex'].value_counts().head(10).plot.pie()
plt.gca().set_aspect('equal')

plt.subplot(2,2,4)
sns.lineplot(data=maths_data['absences'])
plt.xlabel("absences")

```
